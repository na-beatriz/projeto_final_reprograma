# Limpeza e Tratamento dos datasets

```python
#Base de óbitos maternos - Brasil por Cusa Obstetrica e Estado
#Período:2011-2021
#Tratamento da base
arquivos = sorted(glob(r'./datasets/causa_obstetrica/*.csv'))
#Fazer coluna de ano em cada dataset
ano = 2011
inicio = True
for i in arquivos:
    arq_data = pd.read_csv(i, sep=';', encoding='utf-8')
    arq_data.rename(columns={'Unidade da Federa��o':'Estado','Morte materna obst�trica direta':'Morte materna obstétrica direta','Morte materna obst�trica indireta':'Morte materna obstétrica indireta','Morte materna obst�trica n�o especificada':'Morte materna obstétrica não especificada'}, inplace = True)
    df_remove = arq_data.loc[(arq_data['Estado']  == "Total")]
    arq_data = arq_data.drop(df_remove.index)
    arq_data.insert(0,'ANO',allow_duplicates=True,value=ano)
    if inicio:
        df_base_causa_obstetrica = arq_data
        inicio = False
    else:
        df_base_causa_obstetrica = pd.concat([df_base_causa_obstetrica, arq_data],ignore_index=True)
    ano = ano + 1
#Apagar coluna de 'Total'
df_base_causa_obstetrica_2011_2021 = df_base_causa_obstetrica.drop('Total', axis=1)
#Apaga caracteres e espaçamento no início da coluna 'Estado'
df_base_causa_obstetrica_2011_2021['Estado'] = df_base_causa_obstetrica_2011_2021['Estado'].str.slice(3)
#Renomear nome dados da coluna 'Estado'
df_base_causa_obstetrica_2011_2021 = df_base_causa_obstetrica_2011_2021.replace(['Rond�nia','Par�','Amap�','Maranh�o','Piau�','Cear�','Para�ba','Esp�rito Santo','S�o Paulo','Paran�','Goi�s'],['Rondônia','Pará','Amapá','Maranhão','Piauí','Ceará','Paraíba','Espírito Santo','São Paulo','Paraná','Goiás'])
#Exportar para csv
df_base_causa_obstetrica_2011_2021.to_csv('./datasets/bases_tratadas/base_causa_obstetrica_2011_2021.csv')
```

### De acordo com o período que houvem mais falecimentos, quais foram os principais estados das ocorrências?

```python
#Filtragem de todas as ocorrências de 2021
df_ano_falecimento_2021_max = df_obito_mensal_2011_2021[(df_obito_mensal_2011_2021.ANO == 2021) & (df_obito_mensal_2011_2021.Estado)]
#Atualizando o index
df_ano_falecimento_2021_max = df_ano_falecimento_2021_max.reset_index(drop=True)
#Soma de ocorrências por estadp
df_ano_falecimento_2021 = df_ano_falecimento_2021_max.eval('Óbitos = Janeiro + Fevereiro + Março + Abril + Maio + Junho + Julho + Agosto + Setembro + Outubro + Novembro + Dezembro')
#Filtragem das colunas úteis
df_ano_falecimento_2021 = df_ano_falecimento_2021.loc[:,['Estado', 'Óbitos']]
df_ano_falecimento_2021 = df_ano_falecimento_2021.sort_values('Óbitos', ascending=False).reset_index(drop=True)
df_ano_falecimento_2021
```

```python
#Estados com maiores ocorrências de óbitos
df_ano_falecimento_2021_top5_max = df_ano_falecimento_2021.nlargest(5,'Óbitos')
#Atualizando o index
df_ano_falecimento_2021_top5_max = df_ano_falecimento_2021_top5_max.reset_index(drop=True)
df_ano_falecimento_2021_top5_max
```
